```python
from gensim.models import Word2Vec
import nltk
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Restaurant-reviews.csv')
data
```

```python
reviews = data['Review']
reviews
```

```python
sentences_tokens=[]
for  i in range(2000):
       lower_token = []
       token = nltk.word_tokenize(reviews[i])
       for item in token:
              lower_token.append(item.lower())
       sentences_tokens.append(lower_token)

sentences_tokens
```

```python
test = Word2Vec(sentences_tokens, vector_size=100, window=5, min_count=5, workers=4)
```

```python
vector = test.wv['am']
vector
```

```python
word1 ='red'
word2 = 'apple'
similarity = test.wv.similarity(word1, word2)
similarity
```

```python
word = 'red'
embedding = test.wv.get_vector(word)

print(f'The embedding for "{word}" is:\n{embedding}')
```
