This code defines a function called prior_probability() that takes a single input argument, y, 
which is assumed to be a 1-dimensional array of class labels.

The function uses np.unique(y, return_counts=True) to find the unique class labels in y and the number of times 
each class label appears in y.

classes variable will contain the class labels and counts variable will contain the number of occurrences of each class label.

Then it calculates the prior probability of each class label by dividing the number of occurrences of each 
class label by the total number of elements in y using the zip function.

The prior probability is stored as a dictionary in which the keys are the class labels and the values are the prior probabilities.

Finally, the function returns the prior probability dictionary. This function can be used to calculate the 
prior probability of each class label in a given dataset, which can be useful in building a naive Bayes classifier.


```python

def prior_probability(y):
    classes, counts = np.unique(y, return_counts=True)
    prior_prob = dict(zip(classes, counts/len(y)))
    return prior_prob
#
```

```python
def mean_var_by_class(X, y):
    classes, counts = np.unique(y, return_counts=True)
    mean_var = {}
    for cl in classes:
        X_cl = X[y == cl]
        mean_var[cl] = (np.mean(X_cl, axis=0), np.var(X_cl, axis=0))
    return mean_var


```

## 3

```python
def normal_pdf(x, mean, var):
    return (1/(np.sqrt(2*np.pi*var)))*np.exp(-(1/2)*((x-mean)/(var))**2) # var = sigma et mean = mu 
```

##  4

```python
def posterior_probability(X, y, mean_var, prior_prob):
    classes = np.unique(y)
    posterior_prob = {}
    for cl in classes:
        mean, var = mean_var[cl]
        prob_density = normal_pdf(X, mean, var)
        posterior_prob[cl] = np.log(prob_density.prod(axis=1)) + np.log(prior_prob[cl])
    return posterior_prob
```

## 5

```python
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)
```
