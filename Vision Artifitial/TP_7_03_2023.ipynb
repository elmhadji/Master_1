```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read in the input image and convert to grayscale
img = cv2.imread('/home/elmhadji/Work/Python/Vision Artifitial/img/img1.jpg', cv2.IMREAD_GRAYSCALE)

# Get the dimensions of the image
height, width = img.shape[:2]

# Create an output image with the same size as the input image
output = np.zeros((height, width), dtype=np.uint8)

# Prompt the user to enter the threshold values
T1 = int(input('Enter the lowest threshold value: '))
T2 = int(input('Enter the highest threshold value: '))

# Loop over each pixel in the image
for i in range(height):
    for j in range(width):
        # If the pixel intensity is between the two threshold values, set
        # the output image pixel to 255, otherwise set it to 0.
        if T1 < img[i][j] < T2:
            output[i][j] = img[i][j]
        else:
            output[i][j] = 0

# Display the original and thresholded images
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Original Image')
axs[1].imshow(output, cmap='gray')
axs[1].set_title('Thresholded Image')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image in grayscale
img_gray = cv2.imread('/home/elmhadji/Work/Python/Vision Artifitial/img/img1.jpg', cv2.IMREAD_GRAYSCALE)

# Create an empty array to hold the bit planes
bit_planes = []

# Loop over each bit plane from 0 to 7
for k in range(8):
    # Create an array of zeros with the same size as the input image
    bit_plane = np.zeros_like(img_gray)
    
    # Loop over each pixel in the image
    for i in range(img_gray.shape[0]):
        for j in range(img_gray.shape[1]):
            # Get the kth bit of the current pixel value
            b = (img_gray[i, j] >> k) & 1
            # Set the kth bit of the output pixel value to b
            bit_plane[i, j] = (bit_plane[i, j] & ~(1 << k)) | (b << k)
    
    # Add the resulting bit plane to the list
    bit_planes.append(bit_plane)

# Display the results
plt.figure(figsize=(10, 10))
for k in range(8):
    plt.subplot(2, 4, k+1)
    plt.imshow(bit_planes[k], cmap='gray')
    plt.title('Bit Plane {}'.format(k))
    plt.xticks([])
    plt.yticks([])
plt.show()
```
